```python
text = open('gp15.txt','r').read()
wordcloud = WordCloud(max_font_size=100,width = 1520, height = 535, stopwords= STOPWORDS).generate(text)
plt.figure(figsize=(16,9))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
